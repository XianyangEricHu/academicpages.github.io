# Problem 1
Let $X_t$ be an $A R(1)$ process and let $n_t=m t$ for an integer $m>0$ and $t=0,1,2, \ldots$ Define the process $Y_t=X_{n_t}$. Show that $Y_t$ is an $A R(1)$ process.

An AR(1) process is defined by the equation:

$$ X_t = \phi X_{t-1} + \epsilon_t $$

where $ \phi $ is a parameter of the model (with $ |\phi| < 1 $ for stationarity), and $ \epsilon_t $ is a white noise process (i.e., a sequence of uncorrelated random variables with mean 0 and constant variance).

Given $ X_t $ is an AR(1) process and $ Y_t = X_{n_t} $ where $ n_t = mt $ for an integer $ m > 0 $, we need to express $ Y_t $ in terms of $ Y_{t-1} $ to show it is also an AR(1) process.

For $ Y_t = X_{n_t} $, we have $ Y_t = X_{mt} $. Using the AR(1) property of $ X_t $:

$$ X_{mt} = \phi X_{mt-1} + \epsilon_{mt} $$

However, to express this in terms of $ Y_{t-1} $, we need to adjust the indices. Notice that $ mt - 1 $ does not directly translate to $ m(t-1) $ because of the difference created by subtracting 1. To address this, we rewrite $ X_{mt-1} $ in terms of $ X_{m(t-1)} $:

$$ X_{mt-1} = \phi X_{mt-2} + \epsilon_{mt-1} $$
$$ X_{mt-2} = \phi X_{mt-3} + \epsilon_{mt-2} $$
$$ \vdots $$
$$ X_{m(t-1)+1} = \phi X_{m(t-1)} + \epsilon_{m(t-1)+1} $$

Substituting these back into the original equation:
\begin{align*}
X_{mt} &= \phi (\phi X_{mt-2} + \epsilon_{mt-1}) + \epsilon_{mt} \\
       &= \phi^2 X_{mt-2} + \phi \epsilon_{mt-1} + \epsilon_{mt} \\
       & \vdots \\
       &= \phi^m X_{m(t-1)} + \text{terms involving } \epsilon
\end{align*}

Since $ X_{m(t-1)} = Y_{t-1} $, we have:
$$ Y_t = \phi^m Y_{t-1} + \text{noise terms} $$

This shows that $ Y_t $ is also an AR(1) process, with a modified parameter $ \phi^m $ and a different noise structure. The noise terms involve a combination of the $ \epsilon $ terms from the expansion. 

# Problem 2

What is $\mathbb{E}_t\left[\sigma_{t+k}^2\right]$ assuming $\sigma_{t+k}^2$ follows a covariance stationary $\operatorname{GARCH}(1,1)$ process as defined by $(9.35)$ and $(9.42) ? k$ is a non-negative integer. By the expectation at time $t\left(\mathbb{E}_t[]\right)$ we mean the expectation after $y_{t-1}$ and $\sigma_t$ are known but before $y_t$ is known. What is $\lim _{k \rightarrow \infty} \mathbb{E}_t\left[\sigma_{t+k}^2\right]$ ? Your answers should be in terms of the parameters $a_1, b_1, c, k$, and $\sigma_t$.

A GARCH(1,1) process for the conditional variance $\sigma_t^2$ is defined as:

$$ \sigma_{t}^2 = c + a_1 a_{t-1}^2 + b_1 \sigma_{t-1}^2 $$
where:
- $c$ is a constant term,
- $a_1$ and $b_1$ are the GARCH parameters, and
- $a_{t} = \sigma_t\epsilon_{t}$ 

First, let's calculate the unconditional variance for $a_t$.

$$
\begin{aligned}
\operatorname{Var}\left(a_t\right) & =E\left[a_t^2\right]-\left(E\left[a_t\right]\right)^2 \\
& =E\left[a_t^2\right] \\
& =E\left[\sigma_t^2 \epsilon_t^2\right] \\
& =E\left[\sigma_t^2\right] \\
& =c+a_1 E\left[a_{t-1}^2\right]+b_1 \sigma_{t-1}^2 \\
& =c+\left(a_1+b_1\right) E\left[a_{t-1}^2\right]
\end{aligned}
$$
and by stationarity
$$
\begin{aligned}
\operatorname{Var}\left(a_t\right) & =c+\left(a_1+b_1\right) \operatorname{Var}\left(a_t\right) \\
& =\frac{c}{1-a_1-b_1}
\end{aligned}
$$

where
$c, a_1, b_1 \geq 0$ for nonnegativity of variance $a_1+b_1<1$ for stationarity.

For the expectation $\mathbb{E}_t[\sigma_{t+k}^2]$, as $k \rightarrow \infty$, the conditional expectation $\mathbb{E}_t[\sigma_{t+k}^2]$ should converge to the unconditional variance of the process. Given that the process is covariance stationary, the unconditional variance is finite and can be calculated from the parameters of the model. The limit is reached when the effect of the initial condition $\sigma_t^2$ diminishes:

$$ \lim_{k \rightarrow \infty} \mathbb{E}_t[\sigma_{t+k}^2] = \frac{c}{1-a_1-b_1}$$

# Problem 3

After $t$ tosses of a fair coin, let $H(t)$ be the number of heads observed so far; $T(t)$ be the number of tails observed so far; and $X(t)=H(t)-T(t)$.
- (a) Which (if any) of $H(t), T(t)$, and $X(t)$ is an $A R(1)$ process?
- (b) Which (if any) of $H(t), T(t)$, and $X(t)$ is a martingale?
- (c) Which (if any) of $H(t), T(t)$, and $X(t)$ is a Markov Process?
- (d) For what integer values of $n>1$ is $X(t)^n$ a martingale?

## (a)

None of them is an $AR(1)$ process. 

$H(T)\; \text{and}\; T(t)$ are non-decreasing, obviously not $AR(1)$. 

$X(t)$ starts at $0$ and either plus one or minus one, thus a random walk process instead of an $AR(1)$.

## (b)

$H(t)$ and $T(t)$: Neither is a martingale because the expected value of heads (or tails) after the next toss is not equal to the current count. The expectation increases by 0.5 after each toss.

$X(t)$: $X(t)$ is a martingale. Given the current state $X(t)$, the expectation of $X(t+1)$ is $X(t)$ because the expected change (either +1 or -1 with equal probability) is zero.

## (c)

$H(t)$, $T(t)$, and $X(t)$: All of them are Markov processes. The future value of each depends only on its current value and the outcome of the next coin toss, independent of the past sequence of tosses.

## (d)

$X(t)^n$ is a martingale if $\mathbb{E}[X(t+1)^n | F_t)] = X(t)^n$. For a fair coin toss, $X(t+1)$ can either be $X(t)+1$ or $X(t)-1$ with equal probability. We need to find for which values of $n > 1$, this expectation holds.

Consider $n = 2$ (as an example): $\mathbb{E}[X(t+1)^2 | X(t)] = \frac{1}{2} [(X(t)+1)^2 + (X(t)-1)^2] = X(t)^2 + 1$. This is not equal to $X(t)^2$, so $n = 2$ is not a martingale.

For higher values of $n$, the expectation will include terms that depend on $X(t)$ raised to various powers and will not simplify to $X(t)^n$.
Therefore, for integer values of $n > 1$, $X(t)^n$ is not a martingale.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Filter the dataframe to include only call options
call_options = df_spx[df_spx['option_type'] == 'C']

# Convert expiration dates to datetime and find the unique dates
call_options['expiration'] = pd.to_datetime(call_options['expiration'])
unique_expirations = call_options['expiration'].unique()

# Sort the unique expiration dates to find the second oldest
sorted_expirations = sorted(unique_expirations)
second_oldest_expiration = sorted_expirations[1] if len(sorted_expirations) >= 2 else None

# Filter the dataset for the second oldest expiration date
call_options_filtered = call_options[call_options['expiration'] == second_oldest_expiration]

# Calculate the average of the bid and ask prices
call_options_filtered['avg_price'] = (call_options_filtered['bid_eod'] + call_options_filtered['ask_eod']) / 2

# Filter strikes with a delta close to 50, to ensure numerical accuracy
# We approximate this by selecting strikes in increments of 50
min_strike = call_options_filtered['strike'].min()
max_strike = call_options_filtered['strike'].max()
selected_strikes = np.arange(min_strike, max_strike, 50)
call_options_selected = call_options_filtered[call_options_filtered['strike'].isin(selected_strikes)]

# Sort the dataframe based on strike price
call_options_sorted = call_options_selected.sort_values(by='strike')

# Calculate the second derivative of the option price with respect to the strike price
# This is done using numerical differentiation (central difference method)
strikes = call_options_sorted['strike'].values
prices = call_options_sorted['avg_price'].values
second_derivatives = np.gradient(np.gradient(prices, strikes), strikes)

# The second derivative gives the probability density function
pdf_values = second_derivatives

# Ensure that the pdf values are non-negative
pdf_values = np.maximum(pdf_values, 0)

# Plotting the histogram (Implied Probability Mass Function)
plt.figure(figsize=(10, 6))
plt.bar(strikes, pdf_values, width=50, align='center')
plt.title('Implied Probability Mass Function of SPX Prices at Expiration (2022-02-18)')
plt.xlabel('Strike Price')
plt.ylabel('Probability Density')
plt.grid(True)
plt.show()
```

### Observation:

The histogram seems to present a sensible implied probability distribution for the SPX prices at the given expiration date. It captures a market view that includes a central tendency, an expected level of volatility, and a moderate skewness towards higher prices. However, without additional context regarding the market conditions at the time, this interpretation is based solely on the shape of the distribution.

```python
import pandas as pd
from arch import arch_model

# Load the data file
file_path = '/Users/Eric/opt/anaconda3/envs/dsm/3C.csv'
data = pd.read_csv(file_path)
df = data

# Ensure that the 'DEXUSUK' column is numeric and does not contain zeros or negative numbers
df['DEXUSUK'] = pd.to_numeric(df['DEXUSUK'], errors='coerce').dropna()
df = df[df['DEXUSUK'] > 0]

# Convert 'DATE' to datetime and set it as the index
df['Date'] = pd.to_datetime(df['Date'])
df.set_index('Date', inplace=True)

# Calculate the log returns
df['log_returns'] = np.log(df['DEXUSUK']).diff()

# Drop any NaN values that were introduced by the differencing
df = df.dropna()

# Define the training and validation sets
training_set = df.loc['2017-01-01':'2020-12-31', 'log_returns']
validation_set = df.loc['2021-01-01':'2021-12-31', 'log_returns']

# Fit the GARCH(1,1) model
garch_model = arch_model(training_set, p=1, q=1)
garch_result = garch_model.fit(update_freq=5, disp='off')

# Extract the last day of training data for sigma_0 calculation
last_day_of_training = training_set.index[-1]

# Calculate the conditional standard deviation for the last day of training data
sigma_0 = np.sqrt(garch_result.conditional_volatility[last_day_of_training]**2)

# Initialize the sigma list with the sigma_0 value
sigmas = [sigma_0]

# Use the GARCH(1,1) model parameters to calculate sigma for the validation set
params = garch_result.params
for i in range(len(validation_set)):
    new_sigma = np.sqrt(params['omega'] + params['alpha[1]'] * validation_set.iloc[i-1]**2 + params['beta[1]'] * sigmas[-1]**2)
    sigmas.append(new_sigma)

# Drop the initial sigma_0 for the standard deviation calculation
sigmas.pop(0)

# Calculate the normalized returns
normalized_returns = validation_set / np.array(sigmas)

# Calculate variance and excess kurtosis of the normalized returns
normalized_variance = np.var(normalized_returns)
excess_kurtosis = normalized_returns.kurtosis()

print(f'Normalized Variance: {normalized_variance}')
print(f'Excess Kurtosis: {excess_kurtosis}')
```

### Observation

- **Normalized Variance: 0.9010782740451098**
- **Excess Kurtosis: 0.663649781810169**

The normalized variance of the time series $ y_i / \sigma_i $ is 0.901, which is close to 1. This suggests that the GARCH(1,1) model has done a fairly good job of capturing the variance of the log-returns of the British Pound, as the ideal value for a well-modeled series should be close to 1.

The ideal excess kurtosis in a well-fitting GARCH model is 0, indicating a normal distribution of returns.The GARCH model shows an excess kurtosis of 0.6636. This positive value indicates that the distribution of normalized returns is leptokurtic, suggesting fatter tails than a normal distribution. This deviation from 0 implies that the model is somewhat capturing the presence of extreme events or outliers in the data, but it is not perfectly aligned with a normal distribution.

In summary, while the GARCH(1,1) model seems to approximate the variance of the series reasonably well, it may not be capturing the tails and peak of the distribution accurately, which could be important for risk management and derivative pricing purposes. Additional model refinements or the use of a different model that accounts for the leptokurtosis typically observed in financial return series might be necessary.

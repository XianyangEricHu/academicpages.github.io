```python
Query = '''
SELECT s.stock_symbol, m.close_price, m.last_updated_date
FROM market_data m
JOIN stock s ON m.stock_id = s.stock_id
WHERE m.last_updated_date = (SELECT MAX(last_updated_date) FROM market_data)
ORDER BY m.close_price DESC
'''
#Q2
pd.read_sql_query(Query, conn)
```

```python
Query = '''
SELECT tsac_id, tsac_type, tsac_date, tsac_price, tsac_volume, stock_id, trader_id, broker_id, (tsac_price * tsac_volume) AS trading_amount
FROM TRANSACTION
ORDER BY TRADING_AMOUNT DESC
'''
#Q1
pd.read_sql_query(Query, conn)
```

```python
Query = '''
SELECT t.trader_id, tr.first_name, tr.last_name, SUM(t.tsac_price * t.tsac_volume) AS total_amount_traded
FROM transaction t
JOIN trader tr ON t.trader_id = tr.trader_id
GROUP BY t.trader_id, tr.first_name, tr.last_name
ORDER BY total_amount_traded DESC
'''
#Q3
pd.read_sql_query(Query, conn)
```

```python
Query = '''
SELECT company_name, trading_volume
FROM company c
JOIN stock s ON c.company_id = s.company_id
JOIN market_data m ON s.stock_id = m.stock_id
ORDER BY m.trading_volume DESC
LIMIT 1
'''
#Q4
pd.read_sql_query(Query, conn)
```

```python
Query = '''
SELECT b.broker_id, b.first_name, b.last_name, COUNT(*) AS num_transactions
FROM transaction t
JOIN broker b ON t.broker_id = b.broker_id
GROUP BY b.broker_id, b.first_name, b.last_name;
'''
#Q5
pd.read_sql_query(Query, conn)
```
